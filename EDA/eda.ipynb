1. Totales Accesos por Tecnología

```python
data = pd.read_csv('C:\\Users\\User\\OneDrive\\Escritorio\\Datasets Telecomunicaciones\\Datos\\Totales Accesos por Tecnología.csv')
```

```python
columnas_a_graficar = ['ADSL', 'Cablemodem', 'Fibra óptica', 'Wireless', 'Otros']

plt.figure(figsize=(10, 6))

#Mediante este bucle vamos a ir graficando las columnas que seleccione anteriormente.
for columna in columnas_a_graficar:
    plt.plot(data['Año'], data[columna], label=columna)

# Invertimos el eje X porque si lo dejamos como estaba me graficaba desde 2024 hasta 2014.
#plt.gca().invert_xaxis()

# Añadir etiquetas y título
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.title('Conexiones por tipo a lo largo de los años')
plt.legend()

# Mostrar el gráfico
plt.show()
```

En este gráficos de líneas podemos ver como las conexiones de tipo Cablemodem y Fibra óptica han ido incrementandos sus números de usuarios a través de estos diez años; y a su vez visualizamos también la gran caida de ADSL (Banda Ancha) en ese mismo transcurso del tiempo. Esto se puede explicar a las necesidades de estos tiempos de tener un internet con más caudal. Con respecto a Otros y Wireless dentro de todo se han mantenido en el tiempo.

```python
plt.figure(figsize=(12,8))
sns.barplot(x=data['Año'],y=data['Cablemodem'])

# Invertimos el eje X porque si lo dejamos como estaba me graficaba desde 2024 hasta 2014.

plt.title('Cantidad de usuarios')
plt.xlabel('Tiempo')
plt.ylabel('Cantidad')
plt.show()
```

Aqui con más atención se puede ver como cada año a ido incrementando en estos diez año que tenemos en nuestros datos.

2. Totales Accesos por Velocidad

```python
acceso_velocidad=pd.read_csv('C:\\Users\\User\\OneDrive\\Escritorio\\Datasets Telecomunicaciones\\Datos\\Totales Accesos por Velocidad.csv')

```

```python
columnas_a_graficar = ['Hasta 512 kbps','Entre 512 Kbps y 1 Mbps','Entre 1 Mbps y 6 Mbps','Entre 6 Mbps y 10 Mbps','Entre 10 Mbps y 20 Mbps','Entre 20 Mbps y 30 Mbps','Más de 30 Mbps','OTROS']

# Sumar los valores de cada columna
suma_columnas = acceso_velocidad[columnas_a_graficar].sum()

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
suma_columnas.plot(kind='bar')
plt.title('Distribución de velocidades de conexión')
plt.xlabel('Categorías')
plt.ylabel('Total')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

En este gráfico podemos observar a nivel general las velocidades que poseen los usuarios de internet en nuestro país.

```python
# Lista de columnas a graficar
columnas_a_graficar = [
    'Hasta 512 kbps',
    'Entre 512 Kbps y 1 Mbps',
    'Entre 1 Mbps y 6 Mbps',
    'Entre 6 Mbps y 10 Mbps',
    'Entre 10 Mbps y 20 Mbps',
    'Entre 20 Mbps y 30 Mbps',
    'Más de 30 Mbps',
    'OTROS'
]

# Configuración de la figura
plt.figure(figsize=(16, 8))

# Graficar cada columna
for columna in columnas_a_graficar:
    plt.plot(acceso_velocidad['Año'], acceso_velocidad[columna], label=columna, linewidth=2)


# Configuración de etiquetas y título
plt.xlabel('Año')
plt.ylabel('Cantidad')
plt.title('Conexiones por tipo a lo largo de los años')
plt.legend(title='Tipos de conexión')
plt.grid()  # Añadir una cuadrícula para mejor legibilidad

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```

Aquí tenemos un gráfico de líneas que nos demuestran como va variando las velocidades a medidad que pasa el tiempo. El que está en su pico máximo actualmente es Más de 30 Mbps y se puede nortar que desde 2019 hasta la actualidad viene en crecida.Lo contrario es el casi de Entre 1 Mbps y 6 Mbps que desde el año 2014 viene en caída libre.

5. Velocidad por provincia

```python
velocidad=pd.read_csv('C:\\Users\\User\\OneDrive\\Escritorio\\Datasets Telecomunicaciones\\Datos\\Velocidad por provincia.csv')

```

Realizo este diccionario debido a que a la hora de graficar me parece un poco más ordenado agruparlos por sus zonas geográficas.

```python

zonas_geograficas = {
    'Buenos Aires': 'Centro',
    'Córdoba': 'Centro',
    'Santa Fe': 'Centro',
    'Mendoza': 'Cuyo',
    'San Juan': 'Cuyo',
    'San Luis': 'Cuyo',
    'Catamarca': 'Noroeste',
    'Jujuy': 'Noroeste',
    'Salta': 'Noroeste',
    'Santiago del Estero': 'Noroeste',
    'Tucumán': 'Noroeste',
    'La Pampa': 'Patagonia',
    'Neuquén': 'Patagonia',
    'Río Negro': 'Patagonia',
    'Chubut': 'Patagonia',
    'Santa Cruz': 'Patagonia',
    'Tierra del Fuego': 'Patagonia',
    'Corrientes': 'Noreste',
    'Formosa': 'Noreste',
    'Chaco': 'Noreste',
    'Misiones': 'Noreste',
    'Entre Ríos': 'Centro',
    'La Rioja': 'Noroeste'
}

# Agregar una nueva columna 'Zona Geográfica' al DataFrame
velocidad['Zona Geográfica'] = velocidad['Provincia'].map(zonas_geograficas)
```

```python
# Gráfico de líneas agrupado por zonas geográficas
plt.figure(figsize=(10, 6))
sns.lineplot(data=velocidad, x='Año', y='Mbps (Media de bajada)', hue='Zona Geográfica', marker="o")
# Invertimos el eje X porque si lo dejamos como estaba me graficaba desde 2024 hasta 2014.
#plt.gca().invert_xaxis()
plt.title('Evolución de Mbps por Zona Geográfica y Año')
plt.xlabel('Año')
plt.ylabel('Mbps (Media de bajada)')
plt.grid()
plt.xticks(rotation=45)  # Rotar etiquetas del eje X para mejor legibilidad
plt.tight_layout()
plt.show()
```

Se puede observar que las distintas velocidades en las distintas zonas geográficas de Argentina. En muchas zonas como Noreste y Centro han sufrido caidas en sus Mbsps a partir del año 2022. En cuanto a Noroeste este decrecimiento proviene del año 2021. En la Patagonia hubo un estancamiento a cuanto a velocidades; y solamente en la región Cuyo hubo un fuerte crecimiento de Mbps.

```python
df_2024 = velocidad[velocidad['Año'] == 2024]  # Filtrar los datos del año 2024

plt.figure(figsize=(10, 6))
sns.barplot(data=df_2024, x='Provincia', y='Mbps (Media de bajada)', palette='viridis')
plt.title('Mbps por Provincia en el Año 2024')
plt.xlabel('Provincia')
plt.ylabel('Mbps (Media de bajada)')
plt.grid(axis='y')  # Mostrar solo la cuadrícula en el eje Y
plt.xticks(rotation=45)  # Rotar etiquetas para mejor legibilidad
plt.tight_layout()
plt.show()
```

Aquí se puede mostrar mejor los Mbps en el corriente año 2024. En la región Cuyo (y en el país también), la provincia que más se destaca es Mendoza con los Mbps más altos. En cuanto a las demás provincias del país están todos en mismo nivel de Mbps salvo algunas excepciones.
